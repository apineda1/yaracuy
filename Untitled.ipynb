```python
import json
import folium
```

```python
with open('mapa_vnzla.json') as f:
    data = json.load(f)
```

```python
data['features'][0]
```

```python
yaracuy = list()
for i in data['features']:
    if i['properties']['name_1'] == 'Yaracuy':
        yaracuy.append(i)
```
